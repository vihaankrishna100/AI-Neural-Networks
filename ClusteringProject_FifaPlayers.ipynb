Fifa Players Dataset:

This dataset is a list of all the players in Fifa with their ratings and age. This machine learning code will use clustering algorithms to predict the inter_rep column or International Reputation. This column has values 1-5, 1 being that the player does not have much of a impact on their national team, and 5 being the player is one of the best players on the national team.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/content/Fifa_Players_Data - Sheet1 (1).csv")
df.head()
```

```python
#dividing data into features and labels
features = df.drop(["inter_rep"], axis=1)
labels = df.filter(["inter_rep"], axis = 1)
features.head()

```

```python
#training  KMeansmodel
features = features.values
km_model = KMeans(n_clusters = 5)
km_model.fit(features)
```

```python
print(km_model.labels_)
```

```python
plt.scatter(features[:,0],
features[:,1],c=
km_model.labels_,cmap = 'rainbow')

#print the centroids
plt.scatter(km_model.cluster_centers_[:,0],
km_model.cluster_centers_[:,1],
s = 100, c = 'black')
```

```python
# converting categorical labels to numbers
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
labels_num = le.fit_transform(labels)

from sklearn.metrics.cluster import adjusted_rand_score
#adjusted_rand_score(labels_true, labels_pred)
print(" ARI = " , adjusted_rand_score(labels_num, km_model.labels_))
#Evaluating K-Means Results (ARI = adjusted_rand_score)
```

```python
#training KMeans on Kvalues from 1 to 10
loss=[]
for i in range(1, 11):
  km = KMeans(n_clusters = i).fit(features)
  loss.append(km.inertia_)
#printing loss against number of clusters
import matplotlib.pyplot as plt
plt.plot(range(1,11),loss)
plt.title('Finding Optimal Clusters via Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('loss')
plt.show()
```

```python
km_model = KMeans(n_clusters = 3)
km_model.fit(features)
```

```python
plt.scatter(features[:,0],
features[:,1],c=
km_model.labels_,cmap = 'rainbow')

#print the predicted centroids
plt.scatter(km_model.cluster_centers_[:,0],
km_model.cluster_centers_[:,1],
s = 100, c = 'black')
```

```python
le = preprocessing.LabelEncoder()
labels = le.fit_transform(labels)

plt.scatter(features[:,0],
features[:,1], c = labels, cmap = 'rainbow')
```

```python
# converting categorical labels to numbers
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
labels_num = le.fit_transform(labels)

from sklearn.metrics.cluster import adjusted_rand_score
#adjusted_rand_score(labels_true, labels_pred)
print(" ARI = " , adjusted_rand_score(labels_num, km_model.labels_))
#Evaluating K-Means Results (ARI = adjusted_rand_score)
```

*Fields and Target Label*

My dataset was on soccer players, and their Fifa ratings. The different features in my dataset that I used to predict my label, which was their International reputation, was their age,overall_rating,potential,inter_rep,weak_foot(1-5),skill_moves(1-5),crossing,finishing,heading_accuracy,short_passing,volleys,dribbling,long_passing,ball_control,acceleration,sprint_speed,shot_power,stamina,strength,long_shots,interceptions,vision,marking, and sliding_tackle.


*Findings*

1. KMeans Clustering

My initial ARI score that I calculated using the KMeans clustering algorithm was -0.004. This algorithm was done with 5 clusters as there was 5 different potential outputs for my label variable. This score is not good as it means that the model did worse than a model that randomly chooses the number for inter_rep. The code then shows the elbow method was used and it found that the optimum amount of clusters is 3. With this new cluster amount the ARI score showed 0.0033. While this score is still not very good, it is a slight improvement.  

2. Agglomerative (Hierarchical) Clustering

This ARI score for this algorithm was lower than the ARI for the KMeans clustering after the elbow method. The Agglomerative Hierarchical Clustering algorithm had an ARI of -0.008. This suggests that this algorithm did not find the clusters of the dataset, and did not find a pattern.

3. Comparison of ARI Scores

Both methods for this dataset did not do very well. The Kmeans ARI was slightly better after the elbow method, but the Agglomerative algorithm did very poorly, having an ARI of negative. The Kmeans algorithm had a low ARI, but was on the positive end. All in all, this dataset did not have the properties needed for clustering, so the ARI's were very low.

4. Potential Dataset Flaws

This dataset may have a few potential flaws in it. With this dataset there are many features, and not all of them may have had a positive impact on the ARI score. If I use dimensionality reduction on this model it may create a better accuracy score. Another potential flaw is that the international reputation of a player may be very high, but their ratings do not show that. This may be because they are a high, talented player compared to their nation, but not very well compared to the rest of soccer players. This can alter the model's perception of the Internation Reputation score.
