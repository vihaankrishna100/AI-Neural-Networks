Using DATETIME library in Pandas

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("/air_quality_no2_long.csv")

air_quality = air_quality.rename(columns={"date.utc": "datetime"})

air_quality.head()
```

```python
air_quality.city.unique()
```

Gives the unique citiies of the data

```python
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])

air_quality["datetime"]
```

```python
air_quality["datetime"].min(), air_quality["datetime"].max()
```

Gives the start and end time of the data we are working with.


```python
air_quality["datetime"].max() - air_quality["datetime"].min()
```

Gives the difference between max and min time.

```python
air_quality["month"] = air_quality["datetime"].dt.month

air_quality.head()
```

Creates a new column only to compare the month of the time.

```python
air_quality.groupby(
    [air_quality["datetime"].dt.weekday, "location"])["value"].mean()
```

Calculates average of NO2 amount for each week.

```python
fig, axs = plt.subplots(figsize=(12, 4))

air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(
    kind='bar', rot=0, ax=axs
)

plt.xlabel("Hour of the day");  # custom x label using Matplotlib

plt.ylabel("$NO_2 (µg/m^3)$");
```

Creates a graph of the NO2 average in the hour of the day. This averages each hour's NO2 amount and

```python
no_2 = air_quality.pivot(index="datetime", columns="location", values="value")
no_2.index.year, no_2.index.weekday
no_2["2019-05-20":"2019-05-21"].plot();
```

Creates a graph for the days may 20th and may 21st of No2 amounts
