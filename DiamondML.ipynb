```python
import pandas as pd
import numpy as np
import seaborn as sns
```

DIAMONDS DATASET

```python
dim_df = sns.load_dataset("diamonds")
dim_df.head()
```

```python
x = dim_df.drop(['price'],axis = 1)
Y = dim_df['price']
```

```python
x.head()
```

```python
numerical = x.drop(['cut', 'color', 'clarity'], axis = 1)

catagorical = x.filter(['cut', 'color', 'clarity'], axis = 1)

catagorical.head()
```

```python
cat_numerical = pd.get_dummies(catagorical,drop_first=True, dtype = int)
cat_numerical.head()
```

```python
x = pd.concat([numerical, cat_numerical], axis = 1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) #scaling the training set
x_test = sc.transform(x_test) #scaling the test set
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
regressor = lin_reg.fit(x_train, Y_train)
Y_pred = regressor.predict(x_test)
```

```python
from sklearn import metrics
print("Mean Absolute Error : ",
      metrics.mean_absolute_error(Y_test , Y_pred ))
print("Mean Squared Error: ",
      metrics.mean_squared_error(Y_test, Y_pred))
print("Root Mean Squared Error : ",
      np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))

print("R2: ",
      metrics.r2_score(Y_test, Y_pred))
```
